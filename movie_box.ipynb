```python
# 第一步 票房数据抓取
import pymongo
import requests
import time

def times(dt):
    """时间转时间戳"""
    #字符串 转 时间数组
    timeArray = time.strptime(dt, "%Y-%m-%d") #  %H:%M:%S
    #时间数组 转 时间戳
    timestamp = int(time.mktime(timeArray))
    return timestamp

def get_ready(ch='boxoffice',dbname='others'):
    '''数据库调用'''
    global mycol, myclient,myhp
    myclient = pymongo.MongoClient("mongodb://localhost:27017/")
    mydb = myclient[dbname]
    mycol = mydb[ch]
    
get_ready()
timestamp = 1424102400 # 2015-2-17,抓取数据的起始时间
headers = {'User-Agent': 'chrome'}

while True:
    timestamp += 86400 # 加一天
    time_local = time.localtime(timestamp)
    #时间数组 转 新的时间格式(20160505),转成 url 所需的格式
    dt = time.strftime("%Y%m%d",time_local) # 时间
    print(dt, end='')
    url = 'https://box.maoyan.com/promovie/api/box/second.json?beginDate=' + dt
    r = requests.get(url,headers=headers) # 请求接口
    time.sleep(1) #数据量较少便设置时间间隔为 1，没有尝试不限速。
    res = r.json() # 返回 json 文件

    if res['success'] == True and times(res['data']['queryDate'])<=1455638400: # 1455638400 控制抓取范围
        mycol.insert_one(res) # 插入数据
    elif res['success'] == False: 
        print('结束')
        break
    else:
        print('出问题了？')
        break
        
```
